```python
import time
import matplotlib.pyplot as plt

def sum_iterative(N):
    result = 0
    for i in range(1, N+1):
        result += i
    return result

def sum_recursive(N):
    if N == 0:
        return 0
    else:
        return N + sum_recursive(N-1)

# Measure time taken for iterative summation
iterative_times = []
for N in range(1, 1001):
    start_time = time.time()
    sum_iterative(N)
    end_time = time.time()
    iterative_times.append(end_time - start_time)

# Measure time taken for recursive summation
recursive_times = []
for N in range(1, 1001):
    start_time = time.time()
    sum_recursive(N)
    end_time = time.time()
    recursive_times.append(end_time - start_time)

# Plotting
plt.plot(range(1, 1001), iterative_times, label='Iterative')
plt.plot(range(1, 1001), recursive_times, label='Recursive')
plt.xlabel('N')
plt.ylabel('Time (seconds)')
plt.title('Time taken for sum of first N natural numbers')
plt.legend()
plt.show()
```

```python
import random
import matplotlib.pyplot as plt

# Function for linear search
def linear_search(arr, key):
    steps = 0
    for i in range(len(arr)):
        steps += 1
        if arr[i] == key:
            return steps
    return steps

# Function for binary search
def binary_search(arr, key):
    steps = 0
    low = 0
    high = len(arr) - 1
    while low <= high:
        steps += 1
        mid = (low + high) // 2
        if arr[mid] == key:
            return steps
        elif arr[mid] < key:
            low = mid + 1
        else:
            high = mid - 1
    return steps

# Generate random array of 10,000 elements
array = [random.randint(1, 1000) for _ in range(10000)]

# Perform searches and count steps
linear_search_steps = []
binary_search_steps = []

for i in range(5):
    search_key = int(input(f"Enter search key {i + 1}: "))

    # Linear search
    linear_steps = linear_search(array, search_key)
    linear_search_steps.append(linear_steps)

    # Binary search (assuming array is sorted)
    sorted_array = sorted(array)
    binary_steps = binary_search(sorted_array, search_key)
    binary_search_steps.append(binary_steps)

    print("Search Key:", search_key)
    print("Linear Search Steps:", linear_steps)
    print("Binary Search Steps:", binary_steps)

# Plot the number of steps taken by each algorithm for the user-provided search keys
search_numbers = list(range(1, 6))
plt.plot(search_numbers, linear_search_steps, label='Linear Search')
plt.plot(search_numbers, binary_search_steps, label='Binary Search')
plt.xlabel('Search Number')
plt.ylabel('Steps Taken')
plt.title('Steps Taken by Linear and Binary Search')
plt.legend()
plt.show()
```

```python
def string_to_int(s):
    if len(s) == 0:
        return 0
    else:
        digit = int(s[0])
        return digit * (10 ** (len(s) - 1)) + string_to_int(s[1:])

# Test the function
input_string = input("Enter a string of digits: ")
result = string_to_int(input_string)
print("Integer representation:", result)
```

```python
def reverse_string(s):
    if len(s) == 0:
        return s
    else:
        return reverse_string(s[1:]) + s[0]

# Test the function
input_string = input("Enter a string: ")
reversed_string = reverse_string(input_string)
print("Reversed string:", reversed_string)
```

```python
def is_palindrome(s):
    # Base case: if the length of the string is 0 or 1, it's a palindrome
    if len(s) <= 1:
        return True
    # Recursive case: check if the first and last characters are equal,
    # then recursively check the substring excluding the first and last characters
    elif s[0] == s[-1]:
        return is_palindrome(s[1:-1])
    else:
        return False

# Test the function
input_string = input("Enter a string: ")
if is_palindrome(input_string):
    print("The string is a palindrome.")
else:
    print("The string is not a palindrome.")
```
